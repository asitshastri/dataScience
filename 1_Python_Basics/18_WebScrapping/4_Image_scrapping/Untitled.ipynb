```python
soup = bs(response.content,'html.parser')
```

```python
#extracting img tags as a list
img_tags = soup.find_all('img')
del img_tags[0]
print(len(img_tags))
```

```python
images_data_mongo = []
for i in img_tags:
    img_url = i['src']
    img_data = requests.get(img_url).content
    mydict = {"image url":img_url,"data":img_data}
    images_data_mongo.append(mydict)
    with open(os.path.join(dirctory,f"{query}_{img_tags.index(i)}.jpg"),"wb")as f:
        f.write(img_data)
```
