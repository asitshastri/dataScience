```python
"""
Q3. What errors are defined in the ArithmeticError class? Explain any two with 
an example.
"""
"""
Following are the errors defined in ArithematicError Class:
1- FloatingPointError
2- OverflowError
3- ZeroDivisionError
4- DecimalException
"""
#Floating point Error
a=0.6
b=0.7
print(f"this prints: {a+b} insdead of 1.3")

#solution
from decimal import Decimal
a =Decimal("0.6")
b =Decimal("0.7")
print(f"Using decimal we can avoid FloatingPointEroor a+b= {a+b}")
```

```python
#OverflowError:
try:
    n=999.999
    for i in (1,1000):
        n-n**i
except OverflowError as e:
    print(e)
```

```python
#ZeroDivisionError:
try:
    a=5
    b=0
    print(a/0)
except ZeroDivisionError as e:
    print(e)
```

```python
"""
Q4. Why LookupError class is used? Explain with an example KeyError and 
IndexError.
"""
"""
ans:
he LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

You can use LookupError exception class to handle both IndexError and KeyError exception classes.

- LookupError
 --> IndexError
 --> KeyError
"""
#KeyError:
try:
    d = {"key1":[1,2,3],"key2":"asit000"}
    d["key10"]
except KeyError as e:
    print(e)
    
#InderError:
try:
    lst=[1,2,3,4]
    lst[10]
except IndexError as e:
    print("index error: ",e)
```
