```python
import plotly.graph_objects as go
import plotly.express as px
```

### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
fig = px.scatter(df,x='age',y='fare',size='fare')
fig.show()
```

### Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
df = px.data.tips()
fig = px.box(df,x="time", y="total_bill",points='all', color='smoker')
fig.show()
```

### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
df = px.data.tips()
fig = px.histogram(data_frame=df,x="sex",y="total_bill",pattern_shape="smoker",color="day",marginal='violin')
fig.show()
```

### Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

#### Scatter matrix

```python
df = px.data.iris()
fig=px.scatter_matrix(df,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],color='species')
fig.show()
```

### Q5. What is Distplot? Using Plotly express, plot a distplot.
**Distplots**
The distplot figure factory displays a combination of statistical representations of numerical data, such as histogram, kernel density estimation or normal curve, and rug plot.<br>
The distplot can be composed of all or any combination of the following 3 components 

- histogram
- curve: (a) kernel density estimation or (b) normal curve, and
- rug plot
