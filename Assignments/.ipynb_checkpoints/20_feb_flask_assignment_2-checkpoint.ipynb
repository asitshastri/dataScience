### Q1. Explain GET and POST methods.
ans:<br>
**GET Method:**
Send data in unencrypted form to the server. Most common method(has a response bodyi.e server upon receiving the info have to send the response.

**POST Methhod:**
Used to send HTML form data to server. Data received by POST method is not cached by the server.

### Q2. Why is request used in Flask?
ans:<br>
The Flask Request Object is used to perform both sending and receiving operations from a user's browser to the server and process the request data from the server

### Q3. Why is redirect() used in Flask?
ans:<br>
Redirection in Flask can be done by returning a redirect response object, created by invoking the redirect() function. The same method is used for redirecting to other routes on the same application and external websites

### Q4. What are templates in Flask? Why is the render_template() function used?
ans:<br>
render_template is used to generate output from a template file based on the Jinja2 engine that is found in the application's templates folder. Note that render_template is typically imported directly from the flask package instead of from flask. templating .

### Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)

@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
