### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

col = np.random.randint(1,100,50)
size = 10*np.random.randint(1,100,50)
plt.scatter(x,y,c=col,s=size, alpha=0.5, cmap='Accent')
plt.colorbar()

```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.subplot(1,2,1)
plt.title("Box-1")
plt.boxplot(box1)
plt.grid()

plt.subplot(1,2,2)
plt.title("Box-2")
plt.boxplot(box2)
plt.grid()
plt.show()
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
fig = plt.figure(figsize=(10,7))
ax = fig.add_axes([0,0,1,1])
bp = ax.boxplot(data)
plt.grid()
plt.show()
```
