#### **df.iloc vs df.loc**<br>
Both are used for rows and column selection but:
- **df.iloc**: Serach for inbuilt indexes
- **df.loc**: searches for named indexes

**df.iloc**

```python
#row selection using **iloc()**
df.iloc[0:2] #does not consider 2
```

```python
df.iloc[0:2,['PassengerId','Survived','Pclass']]
#Does not work as iloc only understan the inbuilt indexes that are int
```

```python
df.iloc[0:2,[0,1,2]]
```

```python
df.iloc[0:2,0:3]
```

**df.loc**

```python
df.loc[0:2] #considers 2
```

```python
df.loc[0:2,['PassengerId','Survived','Pclass']]
#shows tha tloc can also understan named indexes
```
