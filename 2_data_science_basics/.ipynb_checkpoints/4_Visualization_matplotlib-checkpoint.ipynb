```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

### Random Scatter plot

```python
plt.scatter(x,y)
```

#### Beautifying scatter plot

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y, c='r',alpha=.5, marker='x') #c='31DECE0' can take hexa colour code also, #alpha colotrols intensity of the colour
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
plt.title("Scatter Plot")
plt.grid()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
print(x)
print(y)
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,'--r',)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("Sin Wave Graph")
plt.grid()
```

### Bar Plot

```python
x = list('abcde')
y = np.random.rand(5)
```

```python
plt.figure(figsize=(6,2))
plt.bar(x,y,color='#1DECE0')
plt.xlabel("Character data")
plt.ylabel("Num data")
plt.title("Bar Plot")
plt.grid()
```

**Use barh() method to Transpose of the graph i.e switch x and y axis**

```python
plt.barh(x,y,color='#1DECE0')
```

## Histogram
#### x-axis: data
#### y-axis: Frequency

```python
x = [1,2,3,4,5,5,5,5,6,6,7,7,7,8,8,8,8,1,2]
plt.figure(figsize=(6,4))
plt.hist(x,color='#700EBc')
plt.title("Histogram Plot")
plt.grid()
plt.savefig("resources/firsthistogram.png")
```

## 3D Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
#Fist create a figure inside which we can draw the plot
fig = plt.figure()

#create an axis
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c='#9327EB',alpha = 0.8)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
# plt.zlabel("Z-Axis") #No zlabel methos in matplotlib
plt.title("3D Plot")
plt.show()
```
