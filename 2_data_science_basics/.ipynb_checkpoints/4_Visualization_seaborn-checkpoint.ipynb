### Sepal length vs sepal Width visualisation

```python
sns.scatterplot(x = iris.sepal_length, y=iris.sepal_width)
```

### Sepal length vs Petal length Visulisation

```python
sns.scatterplot(x = iris.sepal_length , y = iris.petal_length)
```

**From the aboove graph we can observe a linaer relation b/w them as sepal length increases petal length also increases**

## Tips dataset

```python
tips = sns.load_dataset('tips')
tips
```

### Scatter Plot

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

observation: Linear relationship b/w total bill and tips paid

#### Extracting head()

```python
tips.head()
```

### Extracting Smoker and Non-smoker

```python
tips['smoker'].value_counts()
```

### Graph seperated by smoker and non-smoker
**hue=** attribute divide the data based on specified column groups

```python
sns.relplot(x = tips.total_bill,y= tips.tip,data=tips, hue='smoker')
```

```python
sns.relplot(x = tips.total_bill,y= tips.tip,data=tips, style='smoker')
```

### Graph seperated based on size column
**style=** attribute works same as hue but instead of diffrenet colour it uses different objects to differentiate b/w different groups of a column

```python
sns.relplot(x = tips.total_bill,y= tips.tip,data=tips, style='size')
```

```python
sns.relplot(x = tips.total_bill,y= tips.tip,data=tips, hue='size')
```

### Total_bill v/s Tips Graph:
- seperated by smokers and non smokers also by time the came at restaurant

```python
sns.relplot(x = tips.total_bill,y= tips.tip,data=tips, hue='time',style='smoker')
```

## Catagorical Plot

```python
sns.catplot(x = 'smoker' , y = 'tip' , data = tips )
```

```python
sns.catplot(x = 'day' , y = 'total_bill' , data = tips )
```

## Joint Plot
Creates a scatter plot b/w the two x and y data set also pair them with their respective histograms

```python
sns.jointplot(x = tips.total_bill , y = tips.tip)
```

## Pair Plot
plot all the relationship b/w the different numerical columns of the dataset

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```
