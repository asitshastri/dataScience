### Merge and Join Operations

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
})
```

```python
data1
```

```python
data2
```

```python
#merge Operation
pd.merge(data1,data2)
```

What merge does is that it extracts the common values b/w both the data set
by default **how=inner**
i.e like a intersection in a ven diagram

**how='left'**

```python
pd.merge(data1,data2,how='left')
```

```python
pd.merge(data1,data2,how='right')
```

```python
pd.merge(data1,data2,how='outer')
```

```python
pd.merge(data1,data2,how='cross')
#it will try to create a combination in which it will combine everythin from row 1 of data1 to every row of data 2 and so on.
```

```python
pd.merge(data1,data2,how='left',on='key1')
```

```python
pd.merge(data1,data2,how='left',left_on='key2',right_on='key4')
```

#### Join

```python
data3 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e']
)
data4 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key22':[4,5,6,7,8],
                      'key33':[3,4,5,6,6]
},index=['a','b','h','i','j']
)
```

```python
data3
```

```python
data4
```

```python
data3.join(data4)
```

**By default:** this join method is doing a left join i.e left is the data3 and right is data4 only common values in right will be shown

```python
data3.join(data4,how='inner')
```

```python
data3.join(data4,how='right')
```

```python
data3.join(data4,how='outer')
```

```python
data3.join(data4,how='cross')
```

##### Now passing date as a string
i.e without using date_range function 

```python
df7 = pd.DataFrame({"date":['2022-06-22','2022-06-20','2022-06-19']})
df7
```

```python
df7.dtypes
```

##### In above case the DataFrame dtype is an Object due to which we might not be able to perform some primitive operations on a dataset file where dates are given in a string format

#### So, converting String date format to a datetime64 format using:   <ins>pd.to_datetime() method 1</ins>

```python
df7['update_date'] = pd.to_datetime(df7['date'])
df7
```

```python
df7.dtypes
```

**Creating new column with year only**

```python
df7['year'] = df7['update_date'].dt.year
df7
```

**Creating new column with month only**

```python
df7['month'] = df7['update_date'].dt.month
df7
```

**Creating new column with day only**

```python
df7['day'] = df7['update_date'].dt.day
df7
```
