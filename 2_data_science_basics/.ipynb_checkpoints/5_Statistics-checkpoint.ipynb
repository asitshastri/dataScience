# Introduction to statstics 
**Statistics**: It is the science of <u>collecting</u>, <u>organuising</u> and <u>analysing the data</u>.

#### What is Data?
> Facts or pieces of information is called a data<br>
> Example Height of the student in class-{175cn,180cm,190cm,.....} and IQ of the student{90,100,120.....}

#### Collecting Data:
> means collecting data through some measurement scales or collecting a variety of data based on some problem statement
#### Organising Data:
>Mean we try to organise the collected data in a perticular way so thata we can visualze it or make any conclusion out of it.<br>
>Main aim of organising data is to **help in Decision Making Process**, 
#### Analysing Data
> As a  data analyst our job to make amazing visualization reports by analysing the data to help help stake holders in making a report.


# Types of Statistics
- 1) **Descriptive Stats**: It consists of <u>organising</u> and <u>summerising</u> the data
- 2) **Infrential Stats**: It consists of using data you have to form  <u>conclusions</u>

### 1) Descriptive Stats:
In this type of statistics we learn following concepts:
> 1- Measure of Centeral Tendencies:[Mean,Median, Mode]<br>
> 2- Measure of Dispersion: [Varience,Standard deviation]<br>
> 3- Histograms,Bar-charts,Pie-charts

### 2) Infrential Stats:
In this type of statistics we use diff. techniques for conclusion,like:
> 1- Z-test<br>
> 2- t-test<br>
> 3- Hypothesis testing<br>
> 4- p value<br>
> 5- significance value
<br>

# QQ-Plot
### Chech Normal Distribution Using QQ-Plot(Quantile-Quantile Plot)

```python
import scipy.stats as stat
import numpy as np
import pylab
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df.head()
```

### Now let's see if every feature(like sepal_length, sepal_width, Petal_length, petal_width, species) belongs to a normal distribution or not

```python
##crating a function that takes df and feature as input and create a plot
def plot_feature(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist="norm",plot=pylab) #probplot()- Calculate quantiles for a probability plot, and optionally show the plot.
```

```python
plot_feature(df,'sepal_width')
```

### Now, using sns to plot histogram instead of matplotlib cuz seaborn has kde feature to smoothen the histogram(sns.histplot(df[feature],kde=True))

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
    
```

```python
plot_data(df,'sepal_width')
```

## Probability Plot(right side)
### it takes df[feature] vs theoritical data(std distribution) plot, If all the points are in conincide with the red colour line then that means the peticular distribution is a normal distribution

```python
plot_data(df,'petal_width') #not a gausian distribution
```

```python
plot_data(df,'sepal_length') #follows gausian distribution
```

```python
plot_data(df,'petal_length') #not a gausian distribution
```

## assignment
### Wap to load tips dataset and plot QQ plot for all its numerical features

```python
tips = sns.load_dataset('tips')
tips.head()
plot_data(tips,'total_bill')
plot_data(tips,'tip')
plot_data(tips,'size') ##not a continous dataset
plot_data(tips,'sex') ##only numerical values allowed
```

# CHI SQUARE TEST PYTHON

```python
import scipy.stats as stat
import numpy as np

## No of hours student study daily in a weekly basis
##monday,tuesday,Wednesday,thursday,Friday,Saturday,sunday
expected_data=[8,6,7,9,6,9,7]
observed_data=[7,8,6,9,9,6,7]

print(f'Expected Data  sum: {sum(expected_data)}, Observed Data sum: {sum(observed_data)}')
##valid data as sum of both must be same


## checking: Chi Square Goodness of fit
chi_sq_test,p_value = stat.chisquare(observed_data,expected_data)
print(f'CHI SQUARE TEST VALUE= {chi_sq_test} and P value= {p_value}')



#Finding critical value
## let significance value be: 0.05
significance_value = 0.05
dof = len(observed_data)-1
critical_value = stat.chi2.ppf(1-significance_value,dof)
print(f'Critical Value calculated using chi2.ppf method= {critical_value}')

if chi_sq_test> critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

# F-Test(Varience Ratio Test) with Python

```python
import scipy.stats as stat
import numpy as np

#Evidence to say that the two population variences are not equal
## check weather onne worker is better than the other one

worker1 = [18,19,22,25,27,28,41,45,51,55]
worker2 = [14,15,15,17,18,22,25,25,27,34]

##calculating F-test statistics
f_test = np.var(worker1)/np.var(worker2)
print(f"f-test/Varience ratio = {f_test}")


##degree of freedom for both workers
df1 = len(worker1)-1
df2 = len(worker2)-2
print(f"df1= {df1}, df2= {df2}")

#significance value
significance_value = 0.05
#confidenceInterval
ci = 1-significance_value

##finding critical value
cv = stat.f.ppf(q =ci, dfn=df1, dfd = df2 )    ##percent point funnction(q=<confidence interval>, dfn<dof of numerator>, dfd<dof of denominator> )
print(f"Critical Value= {cv}")


#Conclusion
if f_test>cv:
    print("Reject the Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```
