```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Plotting

## Plotting with only one axis
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 (etc., depending on the length of the y-points.

```python
y = [3,6,9,20,30,14]
plt.figure(figsize=(6,2))
plt.plot(y)
```

**use: 'o' To plot only the markers**

```python
plt.figure(figsize=(6,2))
plt.plot(y,'o')
```

# Labels

### Setting Font Size Of The Labels
**fontdict** parameter in xlabel(), ylabel(), and title() to set font properties

```python
plt.figure(figsize=(6,2))
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font3 = {'family':'serif','color':'darkred','size':10}
plt.title("Data Plot", fontdict = font1)
plt.xlabel('X-axis', fontdict=font3)
plt.ylabel('Y-axis', fontdict=font3)
```

### Position The Label
 use the **loc** parameter in title() to position the title

```python
plt.figure(figsize=(6,2))
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font3 = {'family':'serif','color':'darkred','size':10}

plt.title("Data Plot", fontdict = font1,loc='left') #'left','right','center'
plt.xlabel('X-axis', fontdict=font3)
plt.ylabel('Y-axis', fontdict=font3)
```

# ---------------------------------------------------------------------------------------------------------------

# Grid

### Specifying which grid line to display
 use the **axis** parameter in the grid() function to specify which grid lines to displa

```python
plt.figure(figsize=(6,2))
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font3 = {'family':'serif','color':'darkred','size':10}

plt.title("Data Plot", fontdict = font1,loc='left') #'left','right','center'
plt.xlabel('X-axis', fontdict=font3)
plt.ylabel('Y-axis', fontdict=font3)
plt.grid(axis='x') #or: axis='y'
```

### Set Line Properties for the Grid
set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).

```python
plt.figure(figsize=(6,2))
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font3 = {'family':'serif','color':'darkred','size':10}

plt.title("Data Plot", fontdict = font1,loc='left') #'left','right','center'
plt.xlabel('X-axis', fontdict=font3)
plt.ylabel('Y-axis', fontdict=font3)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```

# Sub-plot
**subplot() function you can draw multiple plots in one figure:**
- subplot() function takes three arguments that describes the layout of the figure.
-  rows and columns are represented by the first and second argument
- third argument represents the index of the current plot

```python
x = [i for i in range(2,7)]
y= [i for i in range(7,12)]
plt.subplot(1,2,1)
plt.plot(x,y)

x= np.array([2,4,6,8,10])
y= np.array([3,5,7,9,11])
plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

# Scatter

### Colors
**set your own color with the color or the c argument**

#### Colouring each dot

```python
x = np.random.randint(1,20,13)
y = np.random.randint(1,30,13)
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y,c=colors)
plt.show()
```

### ColorMap
- **A colormap is like a list of colors, where each color has a value that ranges from 0 to 100**

```python
pd.options.display.max_rows=200
pd.read_csv('resources/ColorMaps.csv')
```

```python
cols = np.random.randint(1,100,13)
plt.scatter(x,y,c=cols, cmap='Accent')
plt.colorbar()
plt.show()
```

### Different sizes

```python
x = np.random.randint(1,100,100)
y = np.random.randint(1,100,100)
colors = np.random.randint(1,100,100)
sizes = 10 * np.random.randint(1,100,100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Pie-Chart

### Creating Pie Charts

```python
x = np.random.randint(1,20,5)
plt.pie(x)
plt.show()
```

**By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:**

### Labels

```python
x = np.random.randint(1,20,4)
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(x , labels=mylabels)
plt.show()
```

### Start Angle
**the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.**


```python
x = np.random.randint(1,20,4)
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(x , labels=mylabels ,startangle=120)
plt.show()
```

### Explode

```python
x = np.random.randint(1,20,4)
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2,0,0,0]
plt.pie(x , labels=mylabels ,startangle=120, explode=myexplode)
plt.show()
```

### Shadow

```python
x = np.random.randint(1,20,4)
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2,0,0,0]
plt.pie(x , labels=mylabels ,startangle=120, explode=myexplode, shadow=True)
plt.show()
```

### Legend

```python
x = np.random.randint(1,20,4)
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2,0,0,0]
plt.pie(x , labels=mylabels ,startangle=120, explode=myexplode, shadow=True)
plt.legend()
plt.show()
```

# Legend With Header

```python
x = np.random.randint(1,20,4)
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2,0,0,0]
plt.pie(x , labels=mylabels ,startangle=120, explode=myexplode, shadow=True)
plt.legend(title = 'Fruits')
plt.show()
```
