# Introduction to statstics 
**Statistics**: It is the science of <u>collecting</u>, <u>organuising</u> and <u>analysing the data</u>.

#### What is Data?
> Facts or pieces of information is called a data<br>
> Example Height of the student in class-{175cn,180cm,190cm,.....} and IQ of the student{90,100,120.....}

#### Collecting Data:
> means collecting data through some measurement scales or collecting a variety of data based on some problem statement
#### Organising Data:
>Mean we try to organise the collected data in a perticular way so thata we can visualze it or make any conclusion out of it.<br>
>Main aim of organising data is to **help in Decision Making Process**, 
#### Analysing Data
> As a  data analyst our job to make amazing visualization reports by analysing the data to help help stake holders in making a report.


# Types of Statistics
- 1) **Descriptive Stats**: It consists of <u>organising</u> and <u>summerising</u> the data
- 2) **Infrential Stats**: It consists of using data you have to form  <u>conclusions</u>

### 1) Descriptive Stats:
In this type of statistics we learn following concepts:
> 1- Measure of Centeral Tendencies:[Mean,Median, Mode]<br>
> 2- Measure of Dispersion: [Varience,Standard deviation]<br>
> 3- Histograms,Bar-charts,Pie-charts

### 2) Infrential Stats:
In this type of statistics we use diff. techniques for conclusion,like:
> 1- Z-test<br>
> 2- t-test<br>
> 3- Hypothesis testing<br>
> 4- p value<br>
> 5- significance value
