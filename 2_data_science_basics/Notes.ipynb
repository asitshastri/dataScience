```python
import pandas  as pd
#readinga file
df = pd.read_csv("resources/services.csv",header=None)
print(f'datatype: {type(df)}')

# getting ciolumns names
print(f'column names:{df.columns}')

#checking head
# df.head(10)- #by default 5 rows aare shown

#checking tail
# df.tail()

#getting dtype of all the columns
df = pd.read_csv("resources/services.csv")
print(df.dtypes)

##selecting a Pandas Series
print(df['location_id'])
print(type(df['location_id'])) ##Series is a list but aside form a list series also show the indexes

#converting a pandas series to a list
print(list(df['location_id']))

#selecting a column
print(df[['location_id']])

#selecting multiple column
print(df[['location_id','program_id','email']])

##describing data Frame
df.describe()

# selecting columns with datatypes == object
df = pd.read_csv("https://raw.githubusercontent.com/asitshastri/dataScience/main/2_data_science_basics/resources/titanic_data.csv")
print(df.dtypes=='object')
#sleccting columns with datatypes == object
print(df.dtypes[df.dtypes=='object'].index)

#Extracting columns with specific datatypes
print(df[df.dtypes[df.dtypes=='object'].index])

#slicing column
print(df[['PassengerId','Pclass']][4:11])

#adding new column
df['new_col'] = 0
df.head()

#slicing any categorical values from df
print((pd.Categorical(df['Sex'])))
print(pd.Categorical(df['Sex'])[0]) #prints male
#or
print(df['Sex'].unique())

#data filtering
##age above 18
print(df['Age']>18)
##Extracting age column where age>18
print(df[df['Age']>18])
print(f"Length of ppl with age >18: {len(df[df['Age']>18])}")
##extracting names of ppl with fare =0
print(df[df['Fare']==0]['Name'][0:10])

#mean of a column
print(f"Average Age: {df['Age'].mean()}")

#Extract name of person who paid max fair
list(df[df['Fare']==max(df['Fare'])]['Name'])


#iloc: searches inbuilt indices
#syntax: df.iloc[row_start:rowend+1,clm_start:lm_end+1] or df.iloc[0:3,[name1,name2,name3]]
print(df.iloc[0:2,[0,1,2]])
print(df.iloc[0:2,0:3])

#iloc: searches names indices
print(df.loc[0:2])
print(df.loc[0:2,['PassengerId','Survived','Pclass']])
#------------------------------------------------------------------------------


#------------------------------------------------------------------------------
# Advance 
#------------------------------------------------------------------------------

#creating a series from a list
lst = list(df['Name'])
s1 = pd.Series(lst,index=[i for i in range(20,20+len(df))])
print(s1)
##appending a series
s2 = df['Name'][5:10]
print(s1+s2)

#Dropping a column 
df.drop('PassengerId',axis=1) #default axis is 0, axis 1 means it will work vertically
#does not affect orgnal dataframe
#for permanent drop do: "inplace=True"
df.drop('PassengerId',axis=1,inplace=True)

#Dropping a row
df.drop(2,axis=0,inplace = True)
df.head()

#setting named column as index
df.set_index('Name',inplace=True)
df.head()
#Extracting row using loc
print(df.loc['Futrelle, Mrs. Jacques Heath (Lily May Peel)'])
# using iloc
print(df.iloc[2])
#resetting index
df.reset_index(inplace=True)

#converting Dict to data frame
d ={
    'key1':[1,2,3,4],
    'key2':(5,6,7,8),
    'key3':(9,10,11,12),
    'key4':(13,14,15,16)}
df1 = pd.DataFrame(d)
print(df1.head(1))

#reindexing
df1.reindex(['key2','key4','key1','key3']) #do inplace=True for permanently changing the data
print(df1.head())

#dropping rows containing nan values
df.dropna()
#dropping column contauining nan values
df.dropna(axis=1)

#replacing NaN values
df.fillna('empty',inplace=True)
print(df.head())

#grouping by column value
print(df[df['Age']!='empty'].groupby('Pclass').count()['Age'])
print(df[df['Age']!='empty'].groupby('Pclass').sum()['Age'])

#Transposing/rotating(90degree) a DataFrame
print(df.T)

# Concatinate 2 DataFrame
##1- Horizontal concatination (added below-> rows increased)
df1 = df[['Name','Survived','Pclass']][0:5]
df2 = df[['Name','Survived','Pclass']][5:10]
df3 = pd.concat([df1,df2]) #bydefalult- axis=0
##2 Vertical Concatination (added on right side->columns increased)
df4 = pd.concat([df1,df2],axis=1)

#Merg & Join Operation


#Apply method: same as map function
def create_flag(x):
    if x<10:
        return "Cheap Fare"
    elif x>=10 and x<20:
        return "Mid Fare"
    elif x>=20:
        return "High Fare"
df['Fare_Flag'] = df['Fare'].apply(create_flag)
print(df.head())

def test(x):
    if x.dtypes=='object':
        return 'NA'
    else:
        return x.sum()
df1.apply(test , axis=0)

#Applymap() method: applies function to all the elements of the dataframe
df2 = df1[['Survived','Pclass']]
df2.applymap(lambda x:x**2)

#Sorting: 
##1 Sort_indexes: sorts by the index no.
df1.sort_index(ascending = False)

##2 sort_values: sort by a given column
df1.sort_values('Pclass')

#Displaying whole width of data
pd.set_option("display.max_colwidth" , 1000)
df3 = pd.DataFrame({"desc": ["Data Science Masters course is highly curated and uniquely designed according to the latest industry standards. This program instills students the skills essential to knowledge discovery efforts to identify standard, novel, and truly differentiated solutions and decision-making, including skills in managing, querying, analyzing, visualizing, and extracting meaning from extremely large data sets. This trending program provides students with the statistical, mathematical and computational skills needed to meet the large-scale data science challenges of today's professional world. You will learn all the stack required to work in data science industry including cloud infrastructure and real-time industry projects. This course will be taught in Hindi language.","my name is sudh" , "i use to take data science mastes class"]})
print(df3)

#mathematical Operation
mean = df2['Pclass'].mean()
median = df2['Pclass'].median()
mode = df2['Pclass'].mode()
std = df2['Pclass'].std()
minimum = df2['Pclass'].min()
maximum = df2['Pclass'].max()
summation = df2['Pclass'].sum()
varience  = df2['Pclass'].var()
cumsum = df2['Pclass'].cumsum()
# print(f'mean: {mean}, median: {median}, mode: {mode}, std: {std}, Varience: {varience}  min: {minimum}, max: {maximum}, sum: {summation}, Cummulative sum: {cumsum}')

#Windowing Function
# rolling help to scrol thorugh data
df4 = pd.DataFrame({'a':[i for i in range(10)]})
df4 = df4.rolling(window=2).mean() #rolling is taking 2 dataset at a time
print(df4)

#Pandas Date Functionality
date_data = pd.date_range(start = '2022-03-22' ,end = '2022-06-22')
print(date_data)
#convertimg date_data to a data frame
df_date = pd.DataFrame({'date': date_data})
print(df_date)

##Converting date String to Datetime64 format
df7 = pd.DataFrame({"date":['2022-06-22','2022-06-20','2022-06-19']})
print(df7.dtypes) #object type dates
df7['updated_dates'] = pd.to_datetime(df7['date'])
print(df7.dtypes)

##creating 3 seperate column for year, month and date
df7['year'] = df7['updated_dates'].dt.year
df7['month'] = df7['updated_dates'].dt.month
df7['day'] = df7['updated_dates'].dt.day
print(df7)

##timedelta(days =1 , hours = 5 , minutes = 55) Function
timedelta = pd.Timedelta(days=1, hours=5,minutes=30)
df7['update2'] = df7['updated_dates']+timedelta
print(df7)
```
