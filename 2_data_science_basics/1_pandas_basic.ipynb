##### Reading Exel fle

```python
df2 = pd.read_excel('resources/LUSID Excel - Setting up your market data.xlsx')
# if there are multiple sheet then we can use sheetname argument to select a perticular sheet by name by default it is set to none
print(type(df2))
print(f'this is head:\n{df2.head(3)}\n')
print('-------------------------------------------------------------')
print(f'This is tail:\n{df2.tail(2)}\n')
#printing column names as list
print('-------------------------------------------------------------')
print(f'this is column header list:\n{list(df2.columns)}\n')
print('-------------------------------------------------------------')
#selecting a column
print(f"this is column Unnamed: 3:\n{df2['Unnamed: 3']}")
print('-------------------------------------------------------------')
#selecting multiple columns
print(f"this is multiple column selection:\n{df2[['Unnamed: 7' ,'Unnamed: 6']]}")
```
